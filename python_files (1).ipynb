# conditional control statements: if, else, elif

```
# accept 3 subject marks of a student and check whether student got minimum 30 marks in all subjects or not
# if he gets minimum 30 in all subjects the find average marks else print FAIL.
# avg>=75-->A
# avg<75 and avg>=60-->B
# avg<60 and avg>=45-->C
# avg<45 and avg>=30-->D
# else FAIL
x = int(input("print x marks"))
y = int(input("print y marks"))
z = int(input("print z marks"))
if (x>30 and y>30 and z>30): 
  avg = (x+y+z)/3
  if (avg>=75):
    print("GRADE A")
  elif (avg<75 and avg>=60):
    print("GRADE B")
  elif avg<60 and avg>=45:
    print("GRADE C")
  elif avg<45 and avg>=30:
    print("GRADE D")
  else:
    print("FAIL")
else:
  print("FAIL")
```
